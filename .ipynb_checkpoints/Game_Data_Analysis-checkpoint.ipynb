```python
import sqlite3
import pandas as pd
```

```python
sqlite_file = 'sample.sqlite'
```

```python
conn = sqlite3.connect(sqlite_file)
```

```python
query = "SELECT * FROM account_date_session"
```

```python
df = pd.read_sql_query(query, conn)
```

```python
df
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
Q1 = df[(df['date'] >= "2016-01-01") & (df['date'] <= "2016-03-31")]
Q2 = df[(df['date'] >= "2016-04-01") & (df['date'] <= "2016-06-30")]
Q3 = df[(df['date'] >= "2016-07-01") & (df['date'] <= "2016-09-30")]
Q4 = df[(df['date'] >= "2016-10-01") & (df['date'] <= "2016-12-31")]
```

```python
Q1_total_usage_in_hours= round(sum(Q1["session_duration_sec"])/3600,2)
```

```python
Q1_total_usage_in_hours
```

```python
Q2_total_usage_in_hours= round(sum(Q2["session_duration_sec"])/3600,2)
Q3_total_usage_in_hours= round(sum(Q3["session_duration_sec"])/3600,2)
Q4_total_usage_in_hours= round(sum(Q4["session_duration_sec"])/3600,2)
```

```python
import matplotlib.pyplot as plt
```

```python
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
usage_hours = [Q1_total_usage_in_hours, Q2_total_usage_in_hours, Q3_total_usage_in_hours, Q4_total_usage_in_hours]
```

```python
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
usage_hours = [Q1_total_usage_in_hours, Q2_total_usage_in_hours, Q3_total_usage_in_hours, Q4_total_usage_in_hours]

plt.figure(figsize=(10, 6))
bars = plt.bar(quarters, usage_hours, color='skyblue', edgecolor='black')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, round(yval, 2), ha='center', va='bottom')

plt.title('Total Usage in Hours per Quarter')
plt.xlabel('Quarters')
plt.ylabel('Total Usage (in hours)')
plt.grid(axis='y')

plt.show()
```

```python
unique_ids_Q1 = Q1['account_id'].nunique()
unique_ids_Q2 = Q2['account_id'].nunique()
unique_ids_Q3 = Q3['account_id'].nunique()
unique_ids_Q4 = Q4['account_id'].nunique()
```

```python
unique_ids_Q1
```

```python
unique_ids_Q1 = Q1['account_id'].nunique()
unique_ids_Q2 = Q2['account_id'].nunique()
unique_ids_Q3 = Q3['account_id'].nunique()
unique_ids_Q4 = Q4['account_id'].nunique()

quarters = ['Q1', 'Q2', 'Q3', 'Q4']
unique_ids = [unique_ids_Q1, unique_ids_Q2, unique_ids_Q3, unique_ids_Q4]

plt.figure(figsize=(10, 6))
bars = plt.bar(quarters, unique_ids, color='teal', edgecolor='black')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), va='bottom', ha='center')

plt.title('Unique Account IDs per Quarter')
plt.xlabel('Quarters')
plt.ylabel('Number of Active Accounts')
plt.grid(axis='y')

plt.show()
```

```python
daily_account_counts_Q1 = Q1.groupby('date')['account_id'].nunique()
daily_account_counts_Q2 = Q2.groupby('date')['account_id'].nunique()
daily_account_counts_Q3 = Q3.groupby('date')['account_id'].nunique()
daily_account_counts_Q4 = Q4.groupby('date')['account_id'].nunique()
```

```python
total_account_counts_Q1 = daily_account_counts_Q1.sum()
total_account_counts_Q2 = daily_account_counts_Q2.sum()
total_account_counts_Q3 = daily_account_counts_Q3.sum()
total_account_counts_Q4 = daily_account_counts_Q4.sum()
```

```python
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
total_counts = [total_account_counts_Q1, total_account_counts_Q2, total_account_counts_Q3, total_account_counts_Q4]

plt.figure(figsize=(10, 6))
bars = plt.bar(quarters, total_counts, color='green', edgecolor='black')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), va='bottom', ha='center')

plt.title('Daily Active Users per Quarter')
plt.xlabel('Quarters')
plt.ylabel('Number of Entries')
plt.grid(axis='y')

plt.show()
```

```python
df
```

```python
df['created_time'] = pd.to_datetime(df['created_time'])
```

```python
df
```

```python
Q1_iap_purchase = df[(df['created_time'] >= "2016-01-01") & (df['created_time'] <= "2016-03-31")]
Q2_iap_purchase = df[(df['created_time'] >= "2016-04-01") & (df['created_time'] <= "2016-06-30")]
Q3_iap_purchase = df[(df['created_time'] >= "2016-07-01") & (df['created_time'] <= "2016-09-30")]
Q4_iap_purchase = df[(df['created_time'] >= "2016-10-01") & (df['created_time'] <= "2016-12-31")]
```

```python
Q1_earnings= round(sum(Q1_iap_purchase["iap_price_usd_cents"])/100,2)
Q2_earnings= round(sum(Q2_iap_purchase["iap_price_usd_cents"])/100,2)
Q3_earnings= round(sum(Q3_iap_purchase["iap_price_usd_cents"])/100,2)
Q4_earnings= round(sum(Q4_iap_purchase["iap_price_usd_cents"])/100,2)
```

```python
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
usage_hours = [Q1_earnings, Q2_earnings, Q3_earnings, Q4_earnings]
```

```python

plt.figure(figsize=(10, 6))
bars = plt.bar(quarters, usage_hours, color='orange', edgecolor='black')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, round(yval, 2), ha='center', va='bottom')

plt.title('Total earnings per Quarter is US Dollers')
plt.xlabel('Quarters')
plt.ylabel('US Dollers')
plt.grid(axis='y')

plt.show()
```

```python
sqlite_file = 'sample.sqlite'
conn = sqlite3.connect(sqlite_file)
query = "SELECT * FROM account"
df = pd.read_sql_query(query, conn)
```

```python
unique_user_distribution = df.groupby('country_code')['account_id'].nunique()
N = 10 
top_countries_unique = unique_user_distribution.nlargest(N)
other_unique_sum = unique_user_distribution.drop(top_countries_unique.index).sum()
other_unique_series = pd.Series({'Other': other_unique_sum})
top_countries_unique_combined = pd.concat([top_countries_unique, other_unique_series])
fig, ax = plt.subplots(figsize=(12, 7))  
ax.pie(top_countries_unique_combined, labels=top_countries_unique_combined.index, autopct='%1.1f%%', startangle=140, wedgeprops=dict(edgecolor='black'))
ax.set_title('Unique User Distribution by Country - Top {} Countries and Others'.format(N))
plt.show()
```

```python
sqlite_file = 'sample.sqlite'
conn = sqlite3.connect(sqlite_file)
query = "SELECT * FROM iap_purchase"
df_iap = pd.read_sql_query(query, conn)
```

```python
sqlite_file = 'sample.sqlite'
conn = sqlite3.connect(sqlite_file)
query = "SELECT * FROM account"
df_account = pd.read_sql_query(query, conn)
```

```python
df_joined = pd.merge(df_account, df_iap, on='account_id', how='inner')
revenue_distribution = df_joined.groupby('country_code')['iap_price_usd_cents'].sum()
top_revenue_countries = revenue_distribution.nlargest(N)
other_revenue_sum = revenue_distribution.drop(top_revenue_countries.index).sum()
other_revenue_series = pd.Series({'Other': other_revenue_sum})
top_revenue_combined = pd.concat([top_revenue_countries, other_revenue_series])
fig, ax = plt.subplots(figsize=(12, 7))  
ax.pie(top_revenue_combined, labels=top_revenue_combined.index, autopct='%1.1f%%', startangle=140, wedgeprops=dict(edgecolor='black'))
ax.set_title('Revenue Distribution by Country - Top {} Countries and Others'.format(N))
plt.show()
```
